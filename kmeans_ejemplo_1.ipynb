```python
# --LLAMADO DE LIBRERIAS -------------------------------------------------------
import numpy as np                       # MANEJO DE ARRAYS
import matplotlib.pyplot as plt          # GR√ÅFICAS
from sklearn.datasets import make_blobs  # GENERAR DATOS DE EJEMPLO
from sklearn.cluster import KMeans       # ALGORITMO DE AGRUPAMIENTO
# ------------------------------------------------------------------------------

# Generar datos de ejemplo
data, _ = make_blobs(n_samples=300, centers=4, random_state=141)

# Crea el modelo K-Means con 4 cl√∫steres
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)

# Obtiene las etiquetas de los cl√∫steres y los centroides para visualizarlos
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# --VISUALIZACI√ìN DE LOS DATOS Y LOS CENTROIDES---------------------------------

# Ver los datos del color del cluster al que pertenece
plt.scatter(data[:, 0], data[:, 1], c='lightblue')

#  Coordenadas de los puntos "aleatorios"
p = [[-8, -8, 6, 3], [0, 2, 1, -5]]

# Dibujar los centroides
plt.scatter(p[0], p[1], marker='X', s=200, c='black')
plt.title('Clustering with K-Means')  # Pone el t√≠tulo al gr√°fico
plt.show()  # Mostrar gr√°fico
```

```python
# --LLAMADO DE LIBRERIAS -------------------------------------------------------
import numpy as np                       # MANEJO DE ARRAYS
import matplotlib.pyplot as plt          # GR√ÅFICAS
from sklearn.datasets import make_blobs  # GENERAR DATOS DE EJEMPLO
from sklearn.cluster import KMeans       # ALGORITMO DE AGRUPAMIENTO
# ------------------------------------------------------------------------------

# Generar datos de ejemplo
data, _ = make_blobs(n_samples=300, centers=4, random_state=141)

# Crea el modelo K-Means con 4 cl√∫steres
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)

# Obtiene las etiquetas de los cl√∫steres y los centroides para visualizarlos
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# --VISUALIZACI√ìN DE LOS DATOS Y LOS CENTROIDES---------------------------------

# Ver los datos del color del cluster al que pertenece
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap="rainbow")

#  Coordenadas de los puntos "aleatorios"
p = [[-9, -9, 8, 4], [-2, 4, 0, -7]]

# Dibujar los centroides
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='green')
plt.scatter(p[0], p[1], marker='X', s=200, c='black')
plt.title('Clustering with K-Means')  # Pone el t√≠tulo al gr√°fico
plt.show()  # Mostrar gr√°fico
```
